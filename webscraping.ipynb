```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://tap.az/elanlar/elektronika/telefonlar'
response = requests.get(url)
trainings_soup =  BeautifulSoup(response.text, 'html.parser')
# print(car_soup.prettify())
```

```python
trainings_soup
```

```python
product = []
price=[]
link=[]
date=[]


prod = trainings_soup.findAll('div',attrs={"class":"products-name"})
price2 = trainings_soup.findAll('div',attrs={"class":"products-top"})
link2= trainings_soup.findAll('a',attrs={"class":"products-link"})
date2=trainings_soup.findAll('div',attrs={"class":"products-created"})
for (a, b,c,d) in zip(prod, price2,link2,date2):
    product.append(a.text)
    price.append(b.text)
    link.append("tap.az"+c['href'])
    date.append(d.text)
    
```

```python
df=pd.DataFrame({"Mehsul":product,"Qiymet":price,"Link":link,"Yer ve tarix":date})
df
```

```python
df=df.drop(7,axis=0)
```
